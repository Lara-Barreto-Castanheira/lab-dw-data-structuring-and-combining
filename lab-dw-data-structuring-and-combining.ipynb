```python
import pandas as pd

# Load insurance dataset from an online source
url = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file1.csv"
df= pd.read_csv(url)
```

```python
df.head(5)
```

```python
#Column names should be in lower case
df.columns = [col.lower() for col in df.columns] 
#White spaces in column names should be replaced by _
df.columns = [col.replace(" ","_") for col in df.columns]
```

```python
df.rename(columns= {'st': 'state'}, inplace=True)
print(df.columns)
```

```python
df["gender"]=df["gender"].map({"M":"M","Male":"M","female":"F","Femal":"F", "F":"F"})

df ["gender"].value_counts()
```

```python
df["state"]=df["state"].map({"Cali":"California","AZ":"Arizona","WA":"Washington","Oregon":"Oregon","California":"California","Arizona":"Arizona","Nevada":"Nevada","Washington":"Washington"})
df["state"].value_counts()
```

```python
df["education"].unique()
```

```python
df["education"]=df["education"].map({"Bachelors":"Bachelor", 'Master': 'Master','High School or Below':'High School or Below','College':'College', 'Doctor':'Doctor'})
df["education"].value_counts()
```

```python
#In Customer Lifetime Value, delete the % character
df['customer_lifetime_value'] = df['customer_lifetime_value'].str.replace('%',"")
df
```

```python
df["vehicle_class"].unique()
```

```python
#In vehicle class, "Sports Car", "Luxury SUV" and "Luxury Car" could be replaced by "Luxury"
df["vehicle_class"] = df["vehicle_class"].map({"Sports Car":"Luxury", "Luxury SUV":"Luxury", "Luxury Car": "Luxury","Four-Door Car":"Four-Door Car", "Two-Door Car":"Two-Door Car", "Sports Car":"Sports Car" })
df["vehicle_class"].value_counts()
```

```python
#Customer lifetime value should be numeric
df['customer_lifetime_value'] = df['customer_lifetime_value'].astype(float)
df
```

```python
df.dtypes
```

```python
df["number_of_open_complaints"].unique()
```

```python
df["number_of_open_complaints"]=df["number_of_open_complaints"].map({"1/0/00":"0", '1/2/00': '2','1/1/00':'1','1/3/00':'3', '1/4/00':'4', '1/5/00':'5'})
df["number_of_open_complaints"].value_counts()
```

```python
df['number_of_open_complaints'] = df['number_of_open_complaints'].fillna(0)
df['number_of_open_complaints'] = pd.to_numeric(df['number_of_open_complaints'], errors='coerce').astype(int)
df
```

```python
df['number_of_open_complaints'].value_counts()
```

```python
#Identify any columns with null or missing values. 
#Identify how many null values each column has. 
#You can use the isnull() function in pandas to find columns with null values.
df.isnull().sum()
```

```python
df.dropna(inplace=True)
df
```

```python
df.dtypes
```

```python
df['customer_lifetime_value'] = pd.to_numeric (df['customer_lifetime_value'].astype(int))
df['income'] = pd.to_numeric (df['income'].astype(int))
df['monthly_premium_auto'] = pd.to_numeric (df['monthly_premium_auto'].astype(int))
df['total_claim_amount'] = pd.to_numeric (df['total_claim_amount'].astype(int))
df
```

```python
#Use the .duplicated() method to identify any duplicate rows in the dataframe
df.duplicated().sum()  #There is no duplicates
```

```python
#Verify that your strategy has successfully handled the duplicates by checking for duplicates again using .duplicated().
df.duplicated().sum()  #There is no duplicates
```

```python
#after dropping duplicates, reset the index to ensure consistency.
df = df.reset_index(drop=True)
```

Round the total revenue to 2 decimal points.  Analyze the resulting table to draw insights.

```python
pivot_data = data.pivot_table(index='sales_channel', values=['total_claim_amount'], aggfunc='sum')
```

2. Create a pivot table that shows the average customer lifetime value per gender and education level. Analyze the resulting table to draw insights.

```python
#Create a pivot table that shows the average customer lifetime value per gender and education level. 
#Analyze the resulting table to draw insights.

pivot_data_2 = data.pivot_table(index='education', columns='gender', values=['customer_lifetime_value'], aggfunc='mean')
pivot_data_2
```

```python
pivot_data_2= pivot_data_2['customer_lifetime_value'].round(2)
display(pivot_data_2)
```
